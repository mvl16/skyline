```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

codebook 1:Screenshot 2023-04-30 10.44.55 PM.png

```python
#frequency table for H4PE11
df['H4PE11'].value_counts ()
```

```python
#Create new variable 'forgetful_behavior'
df['forgetful_behavior'] = df['H4PE11']
df['forgetful_behavior']. value_counts()
```

```python
#Remove missing values
df['forgetful_behavior'] = df['forgetful_behavior'].loc [ df ['forgetful_behavior'] !=' ']
df['forgetful_behavior'].value_counts()
```

```python
#Remove response of refused and I don't know (6 and 8)
df['forgetful_behavior'] = df['forgetful_behavior'].loc [ df ['forgetful_behavior'] !='6']
df['forgetful_behavior'] = df['forgetful_behavior'].loc [ df ['forgetful_behavior'] !='8']
df['forgetful_behavior'] = df['forgetful_behavior'].loc [ df ['forgetful_behavior'] !='.']
df['forgetful_behavior'].value_counts()
```

```python
#Format names from numbers
df['forgetful_behavior'].replace(['1','2', '3','4','5',], ['strongly agree','agree','neither agree nor disagree','disagree','strongly disagree'] , inplace= True)
df['forgetful_behavior'].value_counts()
```

```python
#Add bar graph
plt.xlabel("Percent")
plt.ylabel("Agree or Disagree")
plt.title("I often forget to put things back in their proper place")
df['forgetful_behavior'].value_counts(normalize = True).plot(kind ='barh', color = '#40916C', width = 0.9,alpha = 0.5, grid = True )
```
